## Theory

Fibre optic cables consist of an inner core made of glass with refractive index $n_i$ and an outer cladding, also made of glass, with refractive index $n_0$, where $n_i \gt n_0$. The cladding allows the light in the core to travel with minimum losses. We can see the diameters of the core and cladding in *fig. 1*. The core and cladding are then surrounded with shielding for durability.

*Figure 1. Core and cladding sizes of single and multimode cables*

![alt text](http://i67.tinypic.com/sfv4i9.png)

*Figure 2. Ray of light travelling through fibre optic cable*

![alt text](http://i64.tinypic.com/28ui3r5.png)

*Figure 2* shows how light propagates through the core of the cable, which follows the condition of total internal reflection, then the angle of reflection is:

$\theta_c = sin^{-1}(\frac{n_i}{n_{air}})$ (1)

The angle of acceptance, $\theta_i = \theta_{max}$, where light entering at greater than $\theta_i$ is only partially reflected and the signal leaks out of the fibre. Using Snell's Law when light travels from air to the fibre:

$sin\theta_i n_{air} = sin\theta_f n_i$ (2)

And from geometry:

$sin\theta_c = sin(90^\circ - \theta_f) = cos\theta_f = \sqrt{1 - sin^2\theta_f}$ (3)

The critical angle $\theta_c$ for total internal reflection at the core/cladding interface is:

$\frac{n_0}{n_i} = sin\theta_c$ (4)

then,

$\frac{n_0}{n_i} = \sqrt{1 - sin^2\theta_f}$ (5)

rearranged:

$sin\theta_i = \frac{1}{n_{air}}\sqrt{n_i^2 - n_0^2}$ (6)

When the angle of acceptance is at maximum, $\theta_i = \theta_{max}$:

$n_{air}sin\theta_{max} = \sqrt{n_i^2 - n_0^2}$ (7)

$n_{air}sin\theta_{max}$ is known as the Numerical Aperture:

$NA = \sqrt{n_i^2 - n_0^2}$ (8)

Number of modes of fibre optic cable can be calculated using:

$V = n_ika\sqrt{2\Delta}$ (9)

where $k = \frac{2\pi}{\lambda}, \Delta = \frac{n_i^2 - n_0^2}{2n_i^2}$, and a is the radius of the core. We can rearrange the equation by substituting NA into $\Delta$ and then into V:

$\Delta = \frac{NA^2}{2n_i^2}$ (10)

$V = ka\sqrt{\frac{2n_i^2(NA)^2}{2n_i^2}} = ka$(NA) (11)

*Figure 3. Path of light rays through single and multimode step-index (and graded index) fibres*

![alt text](http://www.fiberopticshare.com/wp-content/uploads/2016/09/single-mode-fiber-multimode-fiber-1.jpg)

*Fig. 3* shows how light travels through different fibres. In multimode fibres, there are multiple different modes which have slightly different times of propagation. Only one mode propagates for a particular wavelength in a single mode fibre. The  total number of modes for step-index fibres can be calculated by:

$N = \frac{1}{2}V^2$ (12)

### Speed of Light

In a fibre of length L, and c is the speed of light, the total time for light to travel through the fibre is dependent on the refractive index of the ray of light in the core and cladding of the cable:

$t = \frac{Ln_i / n_0}{c / n_i}$ (13)

### Other calculations and errors

The graphs for single and multimode calculations use Gauss' equation as a line of best fit:

$Power = Ae^{\frac{-(x - B)^2}{C^2}}$ (14)

In *fig. 6 and 7*, we used statistics to get the maximum value of the power and applied a %95 power drop, (0.05)(Power$_{max}$), and rearranged equation (14) to get the angle, x, at the power drop:

$x = \sqrt{-C^2*ln(\frac{Power}{A})} + B$ (15)

To get the error on x:

$\sigma_{x} = \sqrt{(\frac{C^2}{2A\sqrt{-\ln \left(\frac{P}{A}\right)}C})^2\sigma_A^2 + \sigma_B^2 + (\frac{C\ln \left(\frac{P}{A}\right)}{\sqrt{-\ln \left(\frac{P}{A}\right)}C})^2\sigma_C^2}$ (16)

x only gives the value of the angle at the drop but not how far it is from the normal. To calculate the angle of acceptance:

$\theta_i = x - (\theta_{L} + \frac{\theta_{R} - \theta_{L}}{2})$ (17)

with error

$\sigma_{\theta_i }= \sqrt{\sigma_x^2 - \frac{1}{2}^2\sigma_{\theta_L^2} - \frac{1}{2}^2\sigma_{\theta_R^2}}$ (18)

The error for the numerical aperture is:

$\sigma_{NA} = \sqrt{cos^2(\theta_{max})\sigma_{\theta_{max}}^2}$ (19)

For the speed of light experiment, I graphed the time vs the length of the cable and used the Least Square Method to get the line of best fit which shows a linear relationship between the time and the length of the cable:

n = 4, m $\equiv $ slope, b $\equiv$ y-intercept, x = time, y = length $\Rightarrow$ y = mx + b

$m = \frac{n \sum_{n=1}^{4} y_n - \sum_{n=1}^{4} x_ny_n  \sum_{n=1}^{4} x_n}{n \sum_{n=1}^{4} x_n^2 - (\sum_{n=1}^{4} x_n)^2}$ (20)

$b = \frac{\sum_{n=1}^{4} x_n^2 \sum_{n=1}^{4} x_ny_n - \sum_{n=1}^{4} x_n  \sum_{n=1}^{4} x_ny_n}{n \sum_{n=1}^{4} x_n^2 - (\sum_{n=1}^{4} x_n)^2}$ (21)

Finally, the error for the number of modes propagating in the fibre, and the error for the step-index calculation are:

$\sigma_V = \sqrt{(-\frac{L}{t^2} \sigma_t)^2 + (\frac{1}{t}\sigma_L)^2}$ (22)

$\sigma_N = V\sigma_V$ (23)

We made the angular vernier as close to the nearest whole degree as possible so that we didn't have to worry about reading the minutes, and recorded the angle for that power in Logger Pro. 

We repeated this for every 1$^\circ$ turning it in the opposite of the initial direction until it read 0W to get the angle of total internal reflection at the other end. We measured the power at every 0.5$^\circ$ when we were close to the peak (90$^\circ$) to get a smoother curve. We used the two angle measurements when the power was 0W to calculate the normal. I graphed the lengths of the cable vs times it took light to travel as a scatter plot then used the sum of least squares method to get a line of best fit. This shows a linear relationship between the length and time, giving the speed of light through a fibre optic cable.

```
from matplotlib import pyplot as plt
import math
import numpy as np
import pandas as pd

length = [4.27, 1008.28, 2009.85, 3013.86]
time = [19, 25.5, 29.2, 33.6]
m = b = 0
# sum of y, sum of x, sum of x^2, (sum of x)^2, sum of xy
sy = sx = sx2 = s2x = sxy = 0

data = []

# data.append(length)
# data.append([0.30, 0.30, 0.30, 0.30])
# data.append(time)
# data.append([0.20, 0.20, 0.20, 0.20])

# sum of least squares method
def slq():
  global m, b, sy, sx, sx2, s2x, sxy
  
  for i in range(4):
    sy += float(length[i])
    sx += float(time[i])
    sx2 += float(time[i]**2)
    sxy += float(time[i])*float(length[i])
       
  s2x = sx**2
  m = (4*sxy - sy*sx)/(4*sx2 - s2x)
  b = (sx2*sy - sx*sxy)/(4*sx2 - s2x)

slq()

# print("%f %f %f %f"% (sy, sx, sx2, sxy))
# print("%f %f"% (m, b))

# times between 19 and 34
x = np.linspace(19, 34, 100)
# apply linear equation with calculated slope and y-intercept
linear = m*x + b

# print("Table 1. Measurements of time over single mode fibre optic cable lengths")
plt.scatter(time, length)
plt.plot(x, linear,'r-')
plt.xlabel("time (μs)")
plt.ylabel("length (m)")
plt.errorbar(time, length, xerr=0.20, yerr=0.30, fmt='o', ecolor='r', capthick=2)
plt.title("\nFigure 8. Time measured for light to travel through each single mode fibre optic cable length")
plt.rcParams['figure.figsize'] = [15, 10]
# pd.DataFrame(data, columns=["1- Cable Length (m)", "2- Cable error (m)", "3- Time measurement (μs)", "4- Time error (μs)"])
```

### Numerical aperture (NA) calculations

Using the Gaussian fit after calculating the %95 power drop:

$Power = Ae^{\frac{-(x - B)^2}{C^2}}$

Rearranging to solve for the angle, x:

$x = \sqrt{-C^2*ln(\frac{Power}{A})} + B$

The error on x:

$\sigma_{x} = \sqrt{(\frac{C^2}{2A\sqrt{-\ln \left(\frac{P}{A}\right)}C})^2\sigma_A^2 + \sigma_B^2 + (\frac{C\ln \left(\frac{P}{A}\right)}{\sqrt{-\ln \left(\frac{P}{A}\right)}C})^2\sigma_C^2}$

To get the angle of acceptance at the %95 power drop:

$\theta_i = x - (\theta_{L} + \frac{\theta_{R} - \theta_{L}}{2})$

with error:

$\sigma_{\theta_i }= \sqrt{\sigma_x^2 - \frac{1}{2}^2\sigma_{\theta_L^2} - \frac{1}{2}^2\sigma_{\theta_R^2}}$

Finally, we can calculate the numerical aperture NA:

$NA = n_{air}sin\theta_{max} = \sqrt{n_i^2 - n_0^2}$

$\sigma_{NA} = \sqrt{cos^2(\theta_{max})\sigma_{\theta_{max}}^2}$

The results of this experiment shows that the single mode fibre optic cable has a smaller inner core, therefore a smaller angle of acceptance, allowing less modes of propagation than the multimode cable. While the single mode cable has less modes of propagation, it allows light to travel faster and for longer distances because there are very little to no reflections inside the cable. 

Multimode cables are good for shorter distances and allow for more modes, or more wavelengths of light to propagate. Multimode cables start to lose more information at longer distances and are slower, but less expensive. Single mode cables are more expensive than multimode cables, but transfer information faster and are good for long distances.

The reason why optic cables are better is because they have less losses than electric cables which work by transferring electrons instead of light. In the speed of light experiment, we saw that light through the fibre optic cable still travels slower than light through a vacuum. This is true because light travels slower in a medium than a vacuum, and also, in a multimode cable, light is bouncing around in the core of the cable.
